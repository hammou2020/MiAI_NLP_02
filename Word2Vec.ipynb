```
from gensim.models import Word2Vec
#training word2vec on 3 sentences
sentence=["Hôm nay chúng ta học Mì AI theo cách Mì ăn liền".lower().split(),
          "Chúng ta cứ học giỏi là có tiền liền".lower().split()]
model = Word2Vec(sentence, min_count=1,size=300,workers=1,seed=42)

# getting word vectors of a word

print(model['mì'])

#printing similarity index
print("Sim=",model.similarity('ai', 'mì'))

#printing similarity index
print("Sim=",model.similarity('học', 'giỏi'))

#printing similarity index
print("Most sim=",model.most_similar('mì'))
```
